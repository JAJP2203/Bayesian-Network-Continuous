```python
import pandas as pd
import numpy as np
```

```python
# Cargar el archivo CSV en un DataFrame
data = pd.read_csv("/Users/tonitojuarez/Downloads/ckd.csv")

# Obtener los nombres de las columnas
column_names = data.columns

# Inicializar listas para almacenar nombres de variables continuas y categóricas
variables_numericas = []
variables_categoricas = []

# Analizar el tipo de dato de cada columna
for column_name in column_names:
    column_type = data[column_name].dtype

#Si el tipo de dato es numérico, agregar a la lista de variables numericas
    if pd.api.types.is_numeric_dtype(column_type):
        variables_numericas.append(column_name)
#Si el tipo de dato no es numérico, agregar a la lista de variables categóricas
    else:
        variables_categoricas.append(column_name)

# Imprimir las listas de variables continuas y categóricas
print("Variables numericas:", variables_numericas)
print("Variables categóricas:", variables_categoricas)
```

```python
#Variables categóricas
df = data[['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane', 'classification']]
df.head()
```

```python
#Variables numéricas
data = data[['id', 'age', 'bp', 'sg', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo','pcv', 'wc', 'rc']]
data.head()
```

```python
data = data[data['rc'] != '\t?']
data['rc'] = data['rc'].astype(float)
data['rc'].unique()
```

```python
data['wc'].replace(['\t6200', '\t8400'], ['6200','8400'], inplace = True)
data = data[data['wc'] != '\t?']
data['wc'] = data['wc'].astype(float)
data['wc'].unique()
```

```python
data['pcv'].replace(['\t43'], ['43'], inplace = True)
data = data[data['pcv'] != '\t?']
data['pcv'] = data['pcv'].astype(float)
data['pcv'].unique()
```

```python
data = data[['age', 'bp', 'sg', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo','pcv','wc','rc']]
data.head()
```

Ahora si ya tenemos la base de datos con todas las columnas tipo float

```python
column_names = data.columns
for column_name in column_names:
    print(data[column_name].dtype)
```

Ya podemos renombrar las variables para entender que son

```python
data = data.rename(columns={'age':'Edad','bp':'Presion', 'sg': 'Gravedad','bgr':'Glucosa','bu':'Urea','sc':'Creatinina','sod':'Sodio','pot':'Potasio','hemo':'Hemoglobina','pcv':'Packed cell volume','wc':'White blood cell count','rc':'Red blood cell count'})
                            
data.head()
```

```python
data.info()
```

Buscas los valores nulos para hacer imputación

```python
valores_nulos = data.isnull().sum()
print(valores_nulos)
data.shape
```

No podemos dejar celdas con valores vacíos, asi que los vamos a imputar con el promedio 

```python
# Rellena los valores nulos de la edad con la media de la columna
media_edad = data['Edad'].mean()
data['Edad'] = data['Edad'].fillna(media_edad)


# Rellena los valores nulos de la presión con la media de la columna
media_presion = data['Presion'].mean()
data['Presion'] = data['Presion'].fillna(media_presion)

# Rellena los valores nulos de la gravedad con la media de la columna
media_gravedad = data['Gravedad'].mean()
data['Gravedad'] = data['Gravedad'].fillna(media_gravedad)

# Rellena los valores nulos de la glucosa con la media de la columna
media_glucosa = data['Glucosa'].mean()
data['Glucosa'] = data['Glucosa'].fillna(media_glucosa)

# Rellena los valores nulos de la urea con la media de la columna
media_urea = data['Urea'].mean()
data['Urea'] = data['Urea'].fillna(media_urea)

# Rellena los valores nulos de la creatinina con la media de la columna
media_creatinina = data['Creatinina'].mean()
data['Creatinina'] = data['Creatinina'].fillna(media_creatinina)

# Rellena los valores nulos del sodio con la media de la columna
media_sodio = data['Sodio'].mean()
data['Sodio'] = data['Sodio'].fillna(media_sodio)

# Rellena los valores nulos del potasio con la media de la columna
media_potasio = data['Potasio'].mean()
data['Potasio'] = data['Potasio'].fillna(media_potasio)

# Rellena los valores nulos de la hemoglobina con la media de la columna
media_hemo = data['Hemoglobina'].mean()
data['Hemoglobina'] = data['Hemoglobina'].fillna(media_hemo)

# Rellena los valores nulos del volumen células con la media de la columna
media_pcv = data['Packed cell volume'].mean()
data['Packed cell volume'] = data['Packed cell volume'].fillna(media_pcv)

# Rellena los valores nulos de las celulas blancas con la media de la columna
media_blanca = data['White blood cell count'].mean()
data['White blood cell count'] = data['White blood cell count'].fillna(media_blanca)

# Rellena los valores nulos de las celulas rojas con la media de la columna
media_rojo = data['Red blood cell count'].mean()
data['Red blood cell count'] = data['Red blood cell count'].fillna(media_rojo)
```
